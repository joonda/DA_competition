## 검색량

* 기존 축제 중 검색량이 가장 많은 것이 무엇일까 ?
* 1차적인 생각 --> 검색량이 많다면, 인기도 많을 것이다

```python
import pandas as pd
```

```python
fest = pd.read_csv('../data/festival_search_data/DM_CLTUR_FSTVL_SCCNT_RESULT_202403.csv')
fest.head()
```

```python
fest.shape
```

```python
fest["ORIGIN_CL_NM"].value_counts()
```

```python
fest1 = fest[['SRCHWRD_NM', 'SCCNT_VALUE']]
fest1
```

```python
fest2 = fest1.groupby(["SRCHWRD_NM"])["SCCNT_VALUE"].sum().to_frame().reset_index()
fest2
```

```python
fest2 = fest2.sort_values(by='SCCNT_VALUE', ascending=False)
fest2
```

```python
fest2.head(20)
```
