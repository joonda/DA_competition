# 전국 문화축제 표준데이터(시군구 행정동 100미터격자코드 포함)

```python
output = glob.glob('./data/2/*.csv')
output
```

```python
df_list = []

for x in output:
    temp = pd.read_csv(x)
    df_list.append(temp)

df1 = pd.concat(df_list, axis=0)
df1.shape
```

```python
df1.head()
```

```python
df1.columns
```

```python
df1.info()
```

## column 정리

```python
cols = ["FCLTY_NM", "CTPRVN_NM", "SIGNGU_NM", "LEGALDONG_NM", "ADSTRD_NM", "RDNMADR_NM", "OPMTN_PLACE_NM",
        "FSTVL_BEGIN_DE", "FSTVL_END_DE"]
```

```python
df1 = df1[cols].copy()
df1.head()
```

## 시작 및 종료 일자 확인 (2023년 데이터 활용키위함)

```python
df1['FSTVL_BEGIN_DE'] = pd.to_datetime(df1['FSTVL_BEGIN_DE'])
df1['FSTVL_END_DE'] = pd.to_datetime(df1['FSTVL_END_DE'])
df1.info()
```

```python
df1["begin_year"] = df1["FSTVL_BEGIN_DE"].dt.year
df1["end_year"] = df1["FSTVL_END_DE"].dt.year
df1
```

* 시작은 2023년만 존재할 수 있으나, 종료는 2023, 2024년 둘다 가능하다 (2023년에 시작했기 때문)
* 나머지 데이터는 삭제 진행 

```python
df1 = df1[(df1['begin_year'] == 2023) | (df1['end_year'] == 2023)].copy()
```

```python
df1.shape
```

```python
df1['begin_year'].unique(), df1['end_year'].unique()
```

```python
df1.drop(df1[df1['begin_year'] == 2022].index, axis=0, inplace=True)
```

```python
df1.drop(df1[df1['begin_year'] == 2024].index, axis=0, inplace=True)
```

```python
df1.shape
```

```python
df1.head()
```

```python
df1[df1['SIGNGU_NM'].isna()]
```

```python
df1.to_csv("./output/전국 문화축제 표준데이터(시군구 행정동 100미터격자코드 포함).csv", encoding="cp949")
```

* 시군구 정보 필요, 축제정보 크롤링하여 데이터를 가져와야할 듯

# 전국 축제 트렌드 지도

```python
output = glob.glob('./data/3/*.csv')
output
```

```python
df_list = []

for x in output:
    temp = pd.read_csv(x)
    df_list.append(temp)

df2 = pd.concat(df_list, axis=0)
df2.shape
```

```python
df2.info()
```

* 시군구 단위로 나눌것이기 때문에, 시군구 데이터가 null값이 아닌 것을 확인

```python
df2.head()
```

## 필요 없는 컬럼 제거

```python
cols = ["FCLTY_NM", "LC_LA", "LC_LO", "CTPRVN_NM", "SIGNGU_NM", "LEGALDONG_NM", "ADSTRD_NM", "RDNMADR_NM"]
```

```python
df2 = df2[cols].copy()
df2.head()
```

```python
df2.drop_duplicates(inplace=True)
```

```python
df2.head(11)
```

```python
df2_list = []

for addr in df2['SIGNGU_NM']:
    for addr2 in sigungu:
        if addr == addr2:
            df2_list.append(df2[df2['SIGNGU_NM'] == addr])
        else:
            pass

df2 = pd.concat(df2_list, axis=0)
df2.shape
```

```python
df2
```

```python
df2.to_csv("./output/전국 축제 트렌드 지도.csv", encoding="cp949", errors='replace')
```
